```python
def imgages(Name_of_img,Number_of_img):

    driver = webdriver.Chrome(r"C:\Users\ganes\Downloads\Internship_assignment\Chrome_Driver\chromedriver_win32\chromedriver.exe")

    driver.get('https://images.google.com/')

    # entering designation as per Question 
    designation = driver.find_element(By.XPATH,"/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/textarea")
    search_input=Name_of_img #  ‘fruits’, ‘cars’ , ‘Machine Learning’, ‘Guitar’, ‘Cakes’
    designation.send_keys(search_input) 

    search = driver.find_element(By.XPATH,'/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/button/div/span')
    search.click()
    
    time.sleep(3)

    # To fetch the url of images
    url=[]
    url_images = driver.find_elements(By.XPATH,'//div[@class="bRMDJf islir"]/img')
    for i in range(Number_of_img):
        url_image = url_images[i].get_attribute('src')
        url.append(url_image)
    
    Name_img =[]
    Name_im = driver.find_elements(By.XPATH,'//div[@class="BxWQUb yLcQMd"]')
    for i in range(Number_of_img):
        Name_i=Name_im[i].text
        Name_img.append(Name_i)
    
    for i in range(len(url)):
        try:
            base64_data = url[i].split(",")[1]
            image_data = base64.b64decode(base64_data)
            file_path = fr"C:\Users\ganes\Downloads\image{i}.jpg"
            with open(file_path, "wb") as file:
                file.write(image_data)
            print(f"Downloaded image {i+1} successfully.")
        except IndexError:
            print(f"Invalid URL format in index {i}: {url[i]}")
        except Exception as e:
            print(f"An error occurred while downloading image {i+1}: {e}")


    list_no = list(range(1,len(url)+1))

    import pandas as pd
    df = pd.DataFrame({'List No':list_no,'Name Images':Name_img,'URL':url})
    return df
```

```python
MAX_COMMENTS = 1000
comment_counter = 0

comments_list = []
upvotes_list = []
timestamps_list = []
Account_name_list=[] 
a='//div[@class=" style-scope ytd-item-section-renderer style-scope ytd-item-section-renderer"]/ytd-comment-thread-renderer/ytd-comment-renderer/div[3]/div[2]/div[1]/div[2]/yt-formatted-string/a'
b='//div[@class=" style-scope ytd-item-section-renderer style-scope ytd-item-section-renderer"]/ytd-comment-thread-renderer/ytd-comment-renderer/div[3]/div[2]/div[1]/div[2]/h3/a/span'

# Scroll down to the bottom of the page
while True:
    # Scroll to the bottom
    driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
    time.sleep(2)

    # Check if new comments were loaded
    comments = driver.find_elements(By.ID,'comment-content')
    new_comments_count = len(comments) - comment_counter

    # Extract comments, upvotes,Account_name and timestamps for the new comments
    for i in range(comment_counter, comment_counter + new_comments_count):
        comment = comments[i].text
        upvote = driver.find_elements(By.ID, "vote-count-left")[i].get_attribute('aria-label')
        timestamp = driver.find_elements(By.XPATH,a)[i].text
        Account_name = driver.find_elements(By.XPATH,b)[i].text
        
        # Append the extracted comment data to the respective lists
        comments_list.append(comment)
        upvotes_list.append(upvote)
        timestamps_list.append(timestamp)
        Account_name_list.append(Account_name)
        
    # Update the comment counter
    comment_counter += new_comments_count

    # Check if the desired number of comments is reached
    if comment_counter >= MAX_COMMENTS:
        break
```

```python
list_no = list(range(1,len(Account_name_list)+1))
```

```python
import pandas as pd
df = pd.DataFrame({'List No':list_no,'Account name list':Account_name_list,'Comments list':comments_list,'Upvotes list':upvotes_list,'Timestamps list':timestamps_list})
df
```
